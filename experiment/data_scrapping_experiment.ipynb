```python
import requests
import pandas as pd
from urllib.parse import urljoin, urlparse
from bs4 import BeautifulSoup
```

```python
BASE_URL = "https://liquipedia.net/mobilelegends/"

HEADERS = {
    "User-Agent": "Mozilla/5.0 (compatible; Bot/1.0; +https://yourdomain.example/)"
}

def get_soup(url):
    response = requests.get(url, headers=HEADERS)
    response.raise_for_status()
    return BeautifulSoup(response.content, "html.parser")

def absolute(href, base=BASE_URL):
    return urljoin(base, href)
```

```python
recent_page = absolute("Recent_Tournament_Results")
try:
    soup = get_soup(recent_page)
    print("Found recent result page (ok):", recent_page)
except Exception as e:
    print("Gagal akses MSC 2025:", e)
```

```python
content = soup.find(id="mw-content-text")
links = set()

if content:
    for a in content.find_all("a", href=True):
        href = a["href"]
        if href.startswith("/mobilelegends/") or href.startswith("/"):
            if ":" in href:
                continue
            full = absolute(href)
            links.add(full)
```

```python
tournament_links = sorted(links)
print(f"Found {len(tournament_links)} A-Tier links")
# for url in tournament_links[:20]:
#     print("-", url)
    
def looks_like_tournament(url):
    p = urlparse(url).path.lower()
    keywords = ["msc/2025", "mpl/", "turkiye", "super_league"]
    return any(k in p for k in keywords)

tournaments = [u for u in tournament_links if looks_like_tournament(u)]
print(len(tournaments))
```
